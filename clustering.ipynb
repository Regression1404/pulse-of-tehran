```python
from arabic_reshaper import reshape
from bidi.algorithm import get_display

all_data = [selected_data1_2, selected_data2_2,
            selected_data3_2,
            selected_data4_2, selected_data5_2, selected_data6_2,
            selected_data7_2, selected_data8_2,
            selected_data9_2,
            selected_data10_2, selected_data11_2, selected_data12_2,
            selected_data13_2, selected_data14_2,
            selected_data15_2, selected_data16_2, selected_data17_2, selected_data18_2,
            selected_data19_2,
            selected_data20_2, selected_data21_2, selected_data22_2]

axis_names = [
    'تهران - پردیس',
    'رباط کریم - تهران',
    'تهران - پاکدشت',
    'تهران - لواسانات',
    'تهران - ساوه (پرند)',
    'تهران - شمال',
    'پردیس - تهران',
    'تهران - رباط کریم',
    'پاکدشت - تهران',
    'لواسانات - تهران',
    'ساوه - تهران (پرند)',
    'شمال - تهران',
    'اسلامشهر - تهران قم',
    'تهران قم - اسلامشهر',
    'آزادگان - اسلامشهر',
    'اسلامشهر - آزادگان',
    'تهران - جاجرود',
    'جاجرود - تهران',
    'تهران - شهریار',
    'شهریار - تهران',
    'تهران - کرج',
    'کرج - تهران'
]

labels = [get_display(reshape(label)) for label in axis_names]
```

```python
import numpy as np
import src.analysis.comparator as comparator

importlib.reload(comparator)

n = len(all_data)

similarity_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == j:
            similarity_matrix[i, j] = 1
        elif i < j:
            dfc = comparator.DataFrameComparer
            dfc = dfc(all_data[i], all_data[j])
            result = dfc.evaluate_similarity()['Pearson Correlation']['total number of vehicles']
            similarity_matrix[i, j] = result
            similarity_matrix[j, i] = result

similarity_matrix
```

```python
distance_matrix = 1 - similarity_matrix

distance_matrix
```

```python
from scipy.spatial.distance import squareform

condensed_dist = squareform(distance_matrix, checks=False)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(condensed_dist, method='average')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
dendrogram(Z, labels=labels, leaf_rotation=70)
plt.title(get_display(reshape("خوشه‌بندی مسیرها بر اساس الگوی ترافیکی")))
plt.xlabel(get_display(reshape("نام محور")))
plt.ylabel(get_display(reshape("فاصله (1 - ضریب همبستگی پیرسون)")))
plt.tight_layout()
plt.show()
```
